```python
X, Y = 9, 23
chess = []
for i in range(X):
  row = list(input())
  chess.append(row)

cnt_paint = []
w, b = ['W', 'B', 'W', 'B', 'W', 'B', 'W', 'B'], ['B', 'W', 'B', 'W', 'B', 'W', 'B', 'W']
white, black = [w, b, w, b, w, b, w, b], [b, w, b, w, b, w, b, w]
isitwhite = chess[0][0] == 'W'

for row in range(X-7):
  paint = 0
  for col in range(Y-7):
    print(row, col)
    print(chess[col:col+8][row:row+8])
    if isitwhite:
      paint += sum([i != j for i, j in zip(chess[col:col+8][row:row+8], white)])
      isitwhite = False
    else:
      paint += sum([i != j for i, j in zip(chess[col:col+8][row:row+8], black)])
      isitwhite = True
    print(paint)
  cnt_paint.append(paint)

print(cnt_paint)
    
```
